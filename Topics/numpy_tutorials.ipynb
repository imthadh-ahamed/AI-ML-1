Import the **Numpy** package

```python
import numpy as np
```

Create 1D array

```python
data_1d = np.array([1, 2, 3, 4, 5])
```

Create 2D array

```python
data_2d = np.array([[1, 2, 3, 4], [5, 6, 7, 8]])
```

Print the Array shape

```python
print(data_1d.shape)
print(data_2d.shape)
```

Reshape the Array

```python
reshaped_arr = data_2d.reshape(4, 2)
reshaped_arr
```

Print the array dimension

```python
print(data_2d.ndim)
```

Array Indexing and slicing

```python
#Indexing
print(data_1d[0])
print(data_1d[3])
print(data_2d[1,2])
print(data_2d[0,3])
```

```python
#Slicing
print(data_1d[:3])
print(data_2d[:, 1:])
```

Array Operations

```python
a = np.array([1, 2, 3, 4])
b = np.array([5, 6, 7, 9])
```

```python
p = np.array([[1, 2, 3, 4], [5, 6, 7, 8]])
q = np.array([[9, 5, 5, 7], [3, 4, 9, 6]])
```

Addition


```python
add = a + b
print("Add: ", add)


addd = p + q
print("Addd:", addd)
```

Multiplication

```python
print("1D Multi: ", a*b)
print("2D Multi: ", p*q)
```

Find the exponential value

```python
exp_a = np.exp(a)
print("exp_a: ", exp_a)

exp_p = np.exp(p)
print("exp_p: ", exp_p)
```

Find the Log value

```python
log_a = np.log(a)
print("log_a: ", log_a)

log_p = np.log(p)
print("log_p: ", log_p)
```

Broadcasting

```python
a = np.array([1, 2, 3])
b = 10
result = a + b
print(result)
```

Array Aggregation

```python
data = np.array([[1, 2, 3], [4, 5, 6]])
print("Summation of the whole array: ", np.sum(data))
print("Column wise summation: ", np.sum(data, axis=0))
print("Row wise mean: ", np.mean(data, axis=1))
```
