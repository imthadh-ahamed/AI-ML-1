```python
import pandas as pd
from apyori import apriori
```

```python
movie_data = pd.read_csv('/content/sample_data/movie_dataset.csv', header=None)
movie_data
```

```python
movie_data.head()
```

```python
movie_data.shape
```

```python
records = []
for i in range(1, 7501):
  records.append([str(movie_data.values[i, j]) for j in range(0, 20)])
```

```python
association_rules1 = apriori(records, min_support=0.001, min_confidence=0.5, min_lift=1, min_length=2)
association_results1 = list(association_rules1)
association_results1
```

```python
association_rules2 = apriori(records, min_support=0.001, min_confidence=0.5, min_lift=1, min_length=2)
association_results2 = list(association_rules2)
association_results2
```

```python
association_rules3 = apriori(records, min_support=0.001, min_confidence=0.5, min_lift=3, min_length=2)
association_results3 = list(association_rules3)
association_results3
```
