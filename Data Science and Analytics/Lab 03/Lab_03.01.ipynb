## Clustering with KMeans Algorithm

```python
# Import necessary libraries
import pandas as pd
import numpy  as np
from matplotlib import pyplot as plt
import seaborn as sns
sns.set()
from sklearn.cluster import KMeans
```

```python
# Import the dataset
country_data = pd.read_csv('/content/sample_data/Countries.csv')
country_data
```

## Plotting the data

```python
country_data.shape
```

```python
x = country_data['Longitude']
y = country_data['Latitude']

plt.scatter(x, y)
plt.show()
```

## Selecting the features

```python
cluster_data = country_data.iloc[:, [1, 2]]
cluster_data
```

## Standardize the variables

```python
from sklearn.preprocessing import StandardScaler
```

```python
a = 'Longitude'
b = 'Latitude'

ss = StandardScaler()
cluster_data = pd.DataFrame(ss.fit_transform(cluster_data), columns=[a, b])
cluster_data
```

## Clustering

```python
kmeans = KMeans(4)
```

```python
kmeans.fit(cluster_data)
```

```python
identified_clusters = kmeans.fit_predict(cluster_data)
identified_clusters
```

## Clustering results

```python
country_data_cpy = country_data.copy()
country_data_cpy
```

```python
country_data.shape
```

```python
country_data_cpy['clusterNo'] = identified_clusters
```

```python
country_data_cpy
```

```python
p = country_data_cpy['Longitude']
q = country_data_cpy['Latitude']
r = country_data_cpy['clusterNo']

plt.scatter(p, q, c=r, cmap='rainbow')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('Clustered Countries')
plt.colorbar(label='Cluster Number')
plt.show()
```

```python
wcss=[]
for i in range(1,11):
    kmeans = KMeans(i)
    kmeans.fit(cluster_data)
    wcss_iter = kmeans.inertia_
    wcss.append(wcss_iter)
```

```python
number_clusters = range(1,11)
plt.plot(number_clusters,wcss)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('Within-cluster Sum of Squares')
```
