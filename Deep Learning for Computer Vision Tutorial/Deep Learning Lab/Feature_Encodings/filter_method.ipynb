```python
import pandas as pd
```

Create datasets and convert the datas as DataFrame

```python
data = {
    'Age' : [25, 30, 20, 40],
    'Income' : [20000, 45000, 100000, 46000],
    'Loan_approval' : [1, 0, 0, 1]
}
df = pd.DataFrame(data)
df
```

Find the correlations

```python
df.corr()
```

```python
correlation = df.corr()['Loan_approval']
correlation
```

Change the absolute values

```python
correlation = correlation.abs()
correlation
```

Sort the values

```python
sorted_feature = correlation.sort_values(ascending=False)
sorted_feature
```

```python
from sklearn.datasets import load_iris
```

Load the pre-defined(load_iris) datas

```python
data = load_iris()
data
```

Using the **SelectKBest** method

```python
from sklearn.feature_selection import SelectKBest, f_classif
```

```python
#x = independent features
#y = dependent features
x, y = data.data, data.target
```

```python
df = pd.DataFrame(data.data, columns = data.feature_names)
df['target'] = data.target
df
```

```python
k_best = SelectKBest(score_func=f_classif, k=2)
```

```python
x_new = k_best.fit_transform(x, y)
```

```python
selected_indices = k_best.get_support(indices=True)
```

```python
selected_features = df.columns[selected_indices]
selected_features
```
