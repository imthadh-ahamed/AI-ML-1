Import the package Library

```python
import pandas as pd
from category_encoders import BinaryEncoder
```

Create some sample datas

```python
data = {
    'Device': ['Laptop', 'TV', 'Smartphone', 'Microwave', 'Tablet'],
    'Portable' :['Yes', 'No', 'Yes', 'No', 'Yes']
}
```

Change the datas as DataFrame

```python
df = pd.DataFrame(data)
df
```

Create an encoder

```python
encoder = BinaryEncoder(cols=['Portable'])
```

Fit and Transform

```python
df_encoder = encoder.fit_transform(df)
df_encoder
```
