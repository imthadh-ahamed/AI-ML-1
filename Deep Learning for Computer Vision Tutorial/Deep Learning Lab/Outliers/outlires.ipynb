Import the package libraries

```python
import numpy as np
import matplotlib.pyplot  as plt
import scipy.stats as stats
```

Sample datas with outliers

```python
data = np.array([100, 160, 116, 10000, 146, 123, 174, 163, 150, 190])
data
```

Calculate the Z-Scores for each data points

```python
z_scores = np.abs(stats.zscore(data))
z_scores
```

Define a threshold for identifying outliers(commonly, a threshold of 3 is used)

```python
threshold = 2
```

Find and print the outliers based on the Z-Score values

```python
outliers = np.where(z_scores > threshold)[0]
print("outliers (Indices) :", outliers)
```

Create an array to specify the color of each point

```python
colors = 'red'
```

Visulize the data with outliers highlighted

```python
plt.figure(figsize=(8, 6))
plt.scatter(np.arange(len(data)), data, c=colors)
plt.xlabel("Data point Index")
plt.ylabel("Sales prices")
plt.title("Outliers detection using Z-Score")
plt.show()
```
