These lines import the necessary libraries for image processing and visualization.<br>
**cv2** is the **OpenCV library** for **computer vision, matplotlib.pyplot** is for plotting, and **numpy** is for numerical operations.


```python
import cv2
from matplotlib import pyplot as plt
import numpy as np
```

This line reads an image file named 'image.jpg' located in the '/content/sample_data/' directory and assigns it to the variable im.

```python
im = cv2.imread('/content/sample_data/image.jpg')
```

This line retrieves the shape of the image im, which represents its dimensions **(height, width, and number of channels)**. <br>The shape of the image is typically represented as a **tuple (height, width, channels)**.

```python
im.shape
```

This line converts the color space of the image from **BGR (Blue, Green, Red)** to **grayscale** using the cv2.cvtColor() function. <br>
**Grayscale** images have only one channel representing the intensity of brightness, unlike color images that have three channels (RGB/BGR).

```python
im = cv2.cvtColor(im, cv2.COLOR_BGR2GRAY)
im.shape
```

These lines define two convolution kernels:
1. **v_kernel** for detecting **vertical edges**.<br>
2. **h_kernel** for detecting **horizontal edges**.<br>
These kernels are **3x3 matrices** used to perform image filtering operations.

```python
v_kernel = np.array([
    [-1, 0, 1],
    [-1, 0, 1],
    [-1, 0, 1]
])

h_kernel = np.array([
    [-1, -1, -1],
    [0, 0, 0],
    [1, 1, 1]
])
```

These lines apply the vertical and horizontal kernels **(v_kernel and h_kernel, respectively)** to the grayscale image im using the **cv2.filter2D()** function.<br>
This function performs a **2D convolution** operation on the image with the specified kernel.

```python
v_image = cv2.filter2D(im, -1, v_kernel)

h_image = cv2.filter2D(im, -1, h_kernel)
```

These lines use **Matplotlib** to visualize the original grayscale image (im) and the images resulting from the vertical edge detection (v_image) and horizontal edge detection (h_image).<br>
**plt.imshow()** is used to **display the images**, and **plt.title()** **adds titles** to the plots for better understanding.<br>
The argument **cmap='gray'** specifies that the colormap used for plotting should interpret the images as grayscale.

```python
plt.imshow(im, cmap='gray')
plt.title('Original Image')

plt.figure()
plt.imshow(v_image, cmap='gray')
plt.title('Vertical Image')

plt.figure()
plt.imshow(h_image, cmap='gray')
plt.title('Horizontal Image')
```
